```python
import pandas as pd
iris_data = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data")
df = iris_data.copy()
```

```python
#Revoming a row from the data: drop() method
#drop() does not remove the given row number from the 
print(df)
a = df.drop(3) #To remove 3rd row.
print(a) #Here in a indeices are not rearranged just 3 is removed so sequence becomes : 0,1,2,4,5...
a = a.drop(0)
a
```

```python
#Now to Use indices as before, after the change position use : index[]
#index[] will give the equivalent valid postion for the required index.
print(a.index[3])
```

```python
#To make the changes in the same data frame: Not by returning make inplace argument = True
a.drop(5 , inplace=True)
a
```

```python
#Printing all positions with indices:
print(a.index)
print("---------------------------------------------")
print(df.index)
```

```python
#Handling NAN:
#Creating sample NAN data:
import numpy as np
df2 = df.copy()
df.iloc[2:4 , 1:3] = np.nan
print(df.head())
df.describe()
```

```python
#Removing rows with NAN values: df.dropna():
df = iris_data.copy()
df.dropna(inplace = True)
print(df.head())
df.reset_index(drop = True , inplace = True)
df
```

```python
#Now putting other data inplace of NAN entries:
import math 
df = df2.copy()
print(df.head())
df.iloc[2:4 , 1:3] = np.nan
print("Printing boolean value to chech nan : ",math.isnan(df.values[2,1]))
df.columns = ['sl' , 'sw' , 'pl' , 'pw' , 'spe']
#Filling sw columns of nan values with mean of sw column and pl columns of nan values with mean of pl:
df.sw.fillna(df.sw.mean() , inplace = True)
df.pl.fillna(df.pl.mean() , inplace = True)
df
```
