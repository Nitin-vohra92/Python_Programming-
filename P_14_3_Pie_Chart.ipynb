```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
sizes = [3,4,6,2]
labels = ['A' , 'B' , 'C' , 'D']
colors = ['pink','red','purple','blue']
plt.pie(sizes , colors = colors , labels = labels , autopct = '%.1f%%') #colors labels and autopct are optional
#autopct can take a function. And in this case just takes the number after the decimal point and make %age.
plt.axis('equal') #To make the graph round
plt.show()
```

```python
#To highlight a specific sections of a graph
sizes = [3,4,6,2]
ex = [0.1 , 0 , 0 , 0]
plt.pie(sizes , explode = ex)
plt.axis('equal')
plt.show()
```

```python
#Changing the starting direction of plotting: , startangle
sizes = [3,4,6,2] 
plt.pie(sizes , startangle = 100)
plt.axis('equal')
plt.show()
```

```python
#Changing the direction of plotting: Default: Counter Clock Wise
sizes = [3,4,6,2] 
plt.pie(sizes , counterclock = False)
plt.axis('equal')
plt.show()
```
