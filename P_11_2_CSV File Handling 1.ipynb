```python
import csv #Library to work with csv files
with open("sampCS.csv") as ob:
    fileData = csv.reader(ob) #csv.reader does not return a list. It returns a csv.reader type object
    #Also above csv reader 2nd Argument : delimiter = ' | ' To split about | and not commas
    #3rd Argument : skipinitialspace = True : To skip initial spaces
    print(type(fileData))
    #Loop to print (by iterating over) fileData:
    for i in fileData:
        print("--------------------------------------------------------------------------------------------------------")#To easily view
        print(i)
        print(type(i)) #every row is different list
    ob.seek(0) #To bring the file pointer to the starting
    twoDList = list(fileData) 
    print "2D List : ",twoDList
    ob.seek(0)
    print("**********************************************************************************************************")
    data2 = csv.DictReader(ob,skipinitialspace=True)#To read csv file a dictionary
    print(data2)
    for i in data2:
        print(i)
        print(type(i)) #All rows are stored as dictionaries. Key : Column Name , Value : value
    
    print("##########################################################################################################")
    ob.seek(0)
    data3 = csv.DictReader(ob,skipinitialspace=True , fieldnames = ('a','b','c'))
    for i in data3:
        print(i)
        print(type(i))
    
    
```
