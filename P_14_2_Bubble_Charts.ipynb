## Bubble Charts:
These are charts with not only points but also a bubble arround the points whch signify some data. 

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
years = [2012,2013,2014,2015,2016,2017]
salary = [12,13,14,17,19,20]
pop = [100,120,180,250,300,370]
```

```python
#Plotting bubble graph: Use scatter()
plt.scatter(years , salary , s=pop , color = 'Blue') # s(list) is the argument to specify the size of the bubble 
plt.show()
```

```python
plt.scatter(years , salary , s=pop , color = ['r','b','g','y','r','b']) #Different colors.
plt.title("Bubble Chart")
plt.show()
```

```python
#Transparency: usse alpha(0 to 1)
plt.scatter(years , salary , s=pop , color = 'Blue' , alpha = 0.2)
plt.show()
```

```python
plt.scatter(years , salary , s=pop , color = 'Blue' ,edgecolor = 'Red') #edge color
plt.show()
```

```python
plt.scatter(years , salary , s=pop , color = 'Green' , marker = '^') #Marker
```
