```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#To plot a 2d Graph use 2 1D arrays of the size using them as xAxis and yAxis.
x = [1,2,3]
y = [2,4,6]

#Command to produce(NOT print) a discrete(Not connected) graph:
plt.scatter(x,y)
#To show:
plt.show()
```

```python
#To plot continuous graph:
plt.plot(x,y,'r--') 
#3rd argument for string to decide : 1st char: colour   2nd char: Type of line
plt.show()
plt.plot(x,y,'go')
plt.show()
plt.plot(x,y,'bp')
plt.show()
plt.plot(x,y,'y^')
plt.show()
plt.plot(x,y,'g-.')
plt.show()
```

```python
#Ploting x^3
x1 = np.array([1,2,3,4])
y = x1**3
plt.plot(x1,y)
plt.show()

x2 = np.arange(0,5,0.001)
y = x2**3
plt.plot(x2,y)
plt.show()
```

```python
#Another way to customising a graph: Arguments:
x2 = np.arange(0,5,0.1) #Increased the step size to see the dots.
y = x2**3
plt.plot(x2,y,color = 'Pink' , linewidth = 10)
plt.show()
plt.plot(x2 , y , color = 'Aqua' , marker = 'p')
```

```python
#Other important Features:
x2 = np.arange(0,5,0.1) #Increased the step size to see the dots.
y = x2**3
plt.plot(x2,y,color = 'Green', linewidth = 5 , label = 'X Cubed')


y2 = x2**2  #X squared
plt.plot(x2,y2,color = 'Blue', linewidth = 5 , label = 'X Squared')

#To change the axis
plt.axis([0,10,0,100]) #take a list of 4(for 2D Graph): [xStart , xEnd , yStart , yEnd]

plt.ylabel('X^3')
plt.xlabel('X')
plt.title('Graph 1')
plt.legend() #To add labels to the graph

plt.grid() #To make a grid

plt.text(3,27,'Woow',fontsize = 15) #To add text to graph at given coordinates. fontsize is optional.

plt.show()
```
