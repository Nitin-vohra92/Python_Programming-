```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = [1,2,3,4,1,4,2,1,1]
xt = [1,2,3,4]
plt.hist(a) #frequency of each number
plt.xticks(xt)
plt.show() 
```

```python
a = [1,2,1,3,5,4,5,5,4,3,3,10,7,6,7,8,8,20,22,11,14]
plt.hist(a,bins = 21,edgecolor = 'black')
plt.xticks(np.arange(0,22,1))
plt.show()
```

```python
#Bar graphs:
years = [2012 , 2013 , 2014 , 2015 , 2016 , 2017]
population = [100,120,180,250,300,370]
plt.bar(years , population , width = 0.8) #Width is optional
plt.xticks(rotation = 40) #To rotate the text to prevent overlaping.
plt.show()
```
