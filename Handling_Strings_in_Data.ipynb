```python
import pandas as pd
col = ['sl' , 'sw' , 'pl' , 'pw' , 'flower name']
iris = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data" , names = col)
df = iris.copy()
```

```python
df.head()
```

```python
#Adding a 6th Column: Gender = 'Female'
df['Gender'] = 'Female'

df.head()
```

```python
#Changing the first entries to 'male':
df.iloc[0:10 , 5] = 'Male'
df_temp = df.copy()
df.head()
```

```python
#Now we want to make all the males = 0 and all the females = 1
#Method 1: apply(): this takes a function as an argument which works on all element of the data set/sliced data set
#Difine a funtion f:
df = df_temp.copy()
def f(s):
    if(s == 'Male'):
        return 0
    return 1
df['Gender'] = df['Gender'].apply(f)
df.head(15)
```

```python
#Method 2: map() : Teks a dictionary and returns the new data frame
df = df_temp.copy()
df['Gender'] = df.Gender.map({'Male' : 0 , 'Female' : 1})
df.head(15)
```
